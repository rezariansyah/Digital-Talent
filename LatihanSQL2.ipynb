```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

# 1.Report the account representative for each customer.

```python
run ("""
SELECT customerNumber,
customerName,
salesRepEmployeeNumber,
concat(e.firstName,' ', e.lastName) as employeeName
From customers a 
Inner Join
employees e on a.salesRepEmployeeNumber = e.employeeNumber
""")
```

# 2.Report total payments for Atelier graphique.

```python
run ("""
SELECT 
customerName,
YEAR(paymentDate)as Tahun,
SUM(amount) AS TotalBayar
From customers
Inner Join
payments Using(customerNumber)
Where customerName = 'Atelier graphique'
Group by Tahun
""")
```

# 3. Report the total payments by date

```python
run("""
SELECT 
    paymentDate,
    SUM(amount) AS Total
FROM
    payments 
WHERE
    paymentDate
GROUP BY paymentDate
""")
```

# 4. Report the products that have not been sold.

```python
run ("""
SELECT productcode, productname 
FROM products
WHERE productcode NOT IN (
SELECT DISTINCT productcode FROM orderdetails);
""")
```

# 5. List the amount paid by each customer.

```python
run("""
SELECT 
    customerName,
    SUM(amount) AS paidAmount
FROM
    customers a
    INNER JOIN
    payments b 
        ON a.customerNumber = b.customerNumber
GROUP BY customerName
""")
```

# 6. How many orders have been placed by Herkku Gifts?

# 7.Who are the employees in Boston?

# 8. Report those payments greater than $100,000. Sort the report so the customer who made the highest payment appears first.


```python
run("""
SELECT 
    customerName,
    CONCAT(contactFirstName,' ',contactLastName) AS Fullname,
    paymentDate, 
    amount
FROM
    customers a
    INNER JOIN
    payments b 
        ON a.customerNumber = b.customerNumber
WHERE amount>100000
ORDER BY amount DESC
""")
```

# 9. List the value of 'On Hold' orders.

```python
run("""
SELECT
    orderNumber,
    orderDate,
    requiredDate,
    `status`,
    comments,
    customerName,
    CONCAT(contactFirstName,' ',contactLastName) AS Fullname
FROM 
    orders o1 
        INNER JOIN
    customers c1 ON o1.customerNumber = c1.customerNumber 
WHERE `status`='On Hold'
""")
```

# 10. Report the number of orders 'On Hold' for each customer.

```python
run("""
SELECT
   customerName,
   COUNT(*) AS statusOnHold 
FROM 
    customers 
        INNER JOIN
    orders USING (customerNumber)
WHERE status='On Hold'
GROUP BY customerNumber
""")
```

# Many To Many Relationship
# 1. List products sold by order date.

```python
run("""
SELECT 
    productName,    
    orderDate
FROM
    ((orderdetails a
    INNER JOIN
    orders b 
        ON a.orderNumber = b.orderNumber)
    INNER JOIN
    products c 
        ON a.productCode = c.productCode)  
ORDER BY orderDate 
""")
```

# 2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.

```python
run("""
SELECT  
    orderNumber, 
    productName,
    orderDate
FROM
    products
    INNER JOIN orderdetails USING (productCode)
    INNER JOIN orders USING (orderNumber)
WHERE productName = '1940 Ford Pickup Truck'
""")
```

# 3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?

```python
run("""
SELECT  
    orderNumber,
    customerName, 
    SUM(quantityOrdered*priceEach) AS Total
FROM
    customers
    INNER JOIN 
        orders USING (customerNumber)
    INNER JOIN 
        orderdetails USING (orderNumber)
GROUP BY 
    orderNumber
HAVING 
    Total>25000
""")
```

# 4. Are there any products that appear on all orders?

```python
run("""
SELECT 
    productCode, 
    COUNT(*) AS Jumlah
FROM 
    orderdetails
GROUP BY 
    productCode
HAVING 
    Jumlah = (SELECT COUNT(*) FROM orders)
""")
```

# 5. List the names of products sold at less than 80% of the MSRP.

```python
run("""
SELECT orderNUmber,productName,priceEach,MSRP
FROM products
    INNER JOIN orderdetails USING (productCode)
    INNER JOIN orders USING (orderNumber)
WHERE priceEach < MSRP
""")
```

# 6. Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)

```python
run("""
SELECT productName,priceEach-buyPrice AS untung, buyPrice
FROM products
INNER JOIN orderdetails USING (productCode)
WHERE priceEach-buyPrice > buyPrice
""")
```

# 7. List the products ordered on a Monday.

```python
run("""
SELECT 
    productName,
    DAYNAME(orderDate) AS Day,
    orderDate
FROM
    ((orderdetails a
    INNER JOIN
    products b
        ON a.productCode = b.productCode)
    INNER JOIN
    orders c
        ON a.orderNumber = c.orderNumber)  
WHERE 
    DAYNAME(orderDate)='Monday'
""")
```

# 8. What is the quantity on hand for products listed on 'On Hold' orders?

```python
run("""
SELECT SUM(quantityOrdered) AS quantitiy
FROM orders 
    INNER JOIN orderdetails USING (orderNumber)
WHERE status='On Hold'
""")
```
