#### Memilih Data berdasarkan NumArticles dan NumMentions

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
# import numpy as np
X = pd.read_csv('Data\DataClean.csv')
df = pd.DataFrame(X, columns = ['NumArticles','NumMentions'])
df
```

#### Mencari kelompok kluster

```python
kmeans = KMeans(n_clusters=5, random_state=0).fit(df)
y =kmeans.labels_
y
```

```python
dfc=pd.DataFrame(y, columns=['Hasil']) 
dfc
```

```python
result = pd.concat([df,dfc], axis=1, sort=False)
result
```

```python
result.loc[118:150]
```

#### mengelompokkan hasil kluster

```python
result['label_kluster'] = result.Hasil.map({0:'Kelompok 0', 1:'Kelompok 1', 2:'Kelompok 2', 3:'Kelompok 3', 4:'Kelompok 4'})
```

```python
result.groupby('label_kluster').count()
```
