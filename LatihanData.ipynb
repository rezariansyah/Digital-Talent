# Deskripsi Data

- Data ini didapat pada website : https://data.go.id/dataset/siswa-sekolah-menengah-atas-yang-putus-sekolah 
- Data ini dikeluarkan oleh Kementerian Pendidikan dan Kebudayaan yang mana bertanggung jawab dalam pelaksanaan, pembimbingan, -dan pengelolaan kegiatan pendidikan dan kebudayaan di Republik Indonesia.
- Data ini berisi mengenai jumlah siswa SMA yang putus sekolah per Provinsi pada tahun 2008 - 2011
- Penjelasan mengenai Variabel pada Dataset ini:
    1. kode_provinsi: Kode Provinsi
    2. nama_provinsi: Nama Provinsi
    3. tahun: Tahun
    4. jumlah_putus_sekolah: Jumlah siswa SMA yang putus sekolah
    5. persen_putus_sekolah: Persentase siswa SMA yang putus sekolah





```python
import pandas as pd
#data_putus_smk = pd.read_csv('data-putus-sekolah-smk-2008-2011.csv', delimiter = ',', encoding = "utf-8")
data_putus_sma = pd.read_csv('data-putus-sekolah-sma-2008-2011.csv', delimiter = ',', encoding = "utf-8")
data_putus_sma.sample(30)
```

```python
s = data_putus_sma.jumlah_putus_sekolah.sum()
mi = data_putus_sma.jumlah_putus_sekolah.min()
ma = data_putus_sma.jumlah_putus_sekolah.max()
data_putus_sma.groupby('tahun').mean()['jumlah_putus_sekolah']

a=data_putus_sma.groupby(['nama_provinsi','tahun']).mean()['jumlah_putus_sekolah']
print('Jumlah Siswa Putus sekolah Pertahun')
print(a)
#print(s,mi,ma)
```

```python
s = data_putus_sma.jumlah_putus_sekolah.sum()
mi = data_putus_sma.jumlah_putus_sekolah.min()
ma = data_putus_sma.jumlah_putus_sekolah.max()

print ('Jumlah Siswa SMA putus Sekolah selama 2008-2011 : ',s)
print ('Minimum Siswa SMA putus Sekolah selama 2008-2011 : ',mi)
print ('Maksimum SMA putus Sekolah selama 2008-2011 : ',ma)
```

```python
ra = data_putus_sma.groupby(['nama_provinsi']).sum()['jumlah_putus_sekolah']
print('Data Putus Sekolah Per Provinsi selama 2008-2011 :'+'\n')
print(ra)

maks = ra.max()
mins = ra.min()

print('Jumlah siswa yang terbanyak putus sekolah :',maks)

print('Jumlah siswa yang paling sedikit putus sekolah :',mins)
```

```python
ax = data_putus_sma.groupby(['tahun','nama_provinsi']).sum()['jumlah_putus_sekolah']
ax = ax.plot(kind='bar', title ="Rata-Rata Berdasarkan Tahun", figsize=(60,30),legend=True, fontsize=24)
ax.set_xlabel("tahun", fontsize=12)
ax.set_ylabel("Jumlah Putus Sekolah", fontsize=12)
```

```python
ax = data_putus_sma.groupby(['nama_provinsi']).sum()['jumlah_putus_sekolah']
ax = ax.plot(kind='bar', title ="Rata-Rata Berdasarkan Tahun", figsize=(10,15),legend=True, fontsize=12)
ax.set_xlabel("Nama Proviinsi", fontsize=12)
ax.set_ylabel("Jumlah Putus Sekolah", fontsize=12)
```
